```python
##**Задача 2.** Рассчитайте средний возраст для посетителей, которые моложе 50 лет
```

```python
sum_age = 0
cnt_age = 0
for i in age_lst:
    if i < 50:
        sum_age += i
        cnt_age += 1
        avg_age = sum_age/cnt_age
print(round(avg_age,2))    
```

```python
##**Задача 3.** Оберните предыдущий код в функцию, которая на вход принимает список и пороговое значение 
```

```python
def avg_age(lst, thresh):
    sum_age = 0
    cnt_age = 0
    for i in lst:
        if i < thresh:
            sum_age += i
            cnt_age += 1
            avg_age = sum_age/cnt_age
    return avg_age
  
```
